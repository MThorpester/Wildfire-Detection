```python
img_count = 1
for index, row in df.iterrows():
    img_count = 1;
    filename = row["Filename"]
    print(filename)
    filepath = os.path.join('D:/Wildfire/HPWren/HPWREN-FIgLib', filename)
    if os.path.exists(filepath) == False:
        continue
    img = cv2.imread(filepath)
    height, width, channels = img.shape
    min_x,min_y,max_x,max_y = [row["MinX"],row["MinY"],row["MaxX"],row["MaxY"]]
    x_diff = max_x - min_x;
    y_diff = max_y - min_y
    print(f'{x_diff}, {y_diff}')
    new_min_x = 0
    new_min_y = 0
    #Center smoke
    center_x = (299 - x_diff) // 2
    center_y = (299 - y_diff) // 2
    #Smoke next to edge
    if (min_x - center_x) <  0:
        new_min_x = min_x // 2 
        new_max_x = new_min_x + 299
        if (min_y - center_y ) < 0:
            new_min_y = min_y // 2
            new_max_y = new_min_y + 299
        else:
            new_min_y = min_y - center_y
            new_max_y = new_min_y + 299
    elif (min_y - center_y) < 0:
        new_min_y = min_y // 2
        nex_max_y = new_min_y + 299
        if (min_x - center_x ) < 0:
            new_min_x = min_x // 2
            new_max_x = new_min_x + 299
        else:
            new_min_x = min_x - center_x
            new_max_x = new_min_x + 299
    elif (min_x - center_x) >= 0:
        new_min_x = min_x - center_x
        new_max_x = new_min_x + 299
        if (min_y - center_y) < 0:
            new_min_y = new_min_y // 2
            new_max_y = new_min_y + 299 
        else:
            new_min_y = min_y - center_y
            new_max_y = new_min_y + 299   
    elif (min_y - center_y) >= 0:
        new_min_y = min_y - center_y
        new_max_y = new_min_y + 299
        if (min_x - center_x) < 0:
            new_min_x = min_x //2
            new_max_x = new_min_x  + 299
        else:
            new_min_x = min_x - center_x
            new_max_x = new_min_x + 299
    print(f"{new_min_x}:{new_max_x}, {new_min_y}:{new_max_y}")
    if (new_max_x - new_min_x) == 299:
        cropped_image = img[new_min_y:new_max_y, new_min_x:new_max_x]
        of = filename.split('.')
        ofile = of[0] + "_centered_" + str(img_count) + "." + of[1] 
        outfile = os.path.join(ofile_dir,ofile)
        img_count += 1
#     print(outfile)
        cv2.imwrite(outfile, cropped_image)
    
        #Flip Centered
        flip_img = cv2.flip(cropped_image, 1)
        ofile = of[0] + "_flipped_" + str(img_count) + "." + of[1] 
        outfile = os.path.join(ofile_dir,ofile)
        cv2.imwrite(outfile,flip_img)
    
    #Shift
    shift_count = 1
    loop_iterator = 1
    while (shift_count <= 5 and  loop_iterator <= 15):
        rand_x = np.random.randint(-120,120)
        rand_y = np.random.randint(-120,120)
        print(f'filename: {filename}, {min_x}, {new_min_x}, {min_y}, {new_min_y} {rand_x} {rand_y}')
        if ((new_min_x + rand_x) < min_x and (new_min_x + rand_x + 299) > max_x and
            (new_min_x + rand_x) >= 0 and
            (new_min_y + rand_y) < min_y and (new_min_y + rand_y + 299) > max_y and
            (new_min_y + rand_y) >= 0):
                print(f'In - {filename}')
                if (new_min_x + rand_x) >=0 :
                    shift_image = img[new_min_y + rand_y:new_min_y + rand_y + 299,new_min_x + rand_x:(new_min_x + rand_x + 299)]
                    heightl, widthl, channelsl = shift_image.shape
                    if widthl == 299:
                        ofile = of[0] + "_shift_" + str(shift_count) + "." + of[1] 
                        outfile = os.path.join(ofile_dir,ofile)
                        cv2.imwrite(outfile,shift_image)
                        #FLip shifted image
                        flip_img = cv2.flip(shift_image, 1)
                        ofile = of[0] + "_shift_flipped_" + str(shift_count) + "." + of[1] 
                        outfile = os.path.join(ofile_dir,ofile)
                        cv2.imwrite(outfile,flip_img)
                        shift_count += 1
        #Use loop iterator io prevent issues with edge smoke images.
        loop_iterator += 1
```
